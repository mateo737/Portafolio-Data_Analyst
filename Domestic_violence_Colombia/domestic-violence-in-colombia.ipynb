## Domestic violence in Colombia
Domestic violence in Colombia from 2010 to 2021

About Dataset
Context
This is a compilation of data on domestic violence that has been developing in Colombia over time.

Content
It is divided between

Departments: which are political and administrative divisions that make up the national territory.

Municipalities: Which correspond to the second level of administrative division in Colombia

DANE Code: It is a standardized nomenclature, designed by DANE for the identification of Territorial Entities (departments, districts and municipalities), Non-Municipalized Areas and Populated Centers, by assigning a unique numerical code to each of these territorial units.

Weapons Medium: Which is an allowance for the items that were used to commit the act of violence
Date Made: Which describes the day on which the violent event took place

Gender: Sex of the victim
Quantity: Number of people who have been injured

```python
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
# Cargar los datos
df = pd.read_csv("/kaggle/input/domestic-violence-in-colombia/Reporte_Delito_Violencia_Intrafamiliar_Polic_a_Nacional.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
# Convierte la columna 'FECHA HECHO' a tipo datetime
df["FECHA HECHO"] = pd.to_datetime(df["FECHA HECHO"], format='%d/%m/%Y', errors="coerce")

df.dtypes
```

```python
df["FECHA HECHO"].min(),df["FECHA HECHO"].max()
```

## Datos nullos

```python
df.isnull().sum()
```

```python
# Reemplaza los valores nulos en 'GRUPO ETARIO' con 'NO REPORTA'
df['GRUPO ETARIO'].fillna('NO REPORTA', inplace=True)
# Eliminamos los datos nulos restantes
df = df.dropna()
# Remplazar los datos "-" por "NO REPORTADO"
df['ARMAS MEDIOS'].replace('-', 'NO REPORTADO', inplace=True)

df.isnull().sum()
```

```python
df.info()
```

Realizaremos el analisis con 476959 casos. Desde 2010-01-01 a 2021-05-31.

## Analisis departamental y municipal

```python
top_dep = df["DEPARTAMENTO"].value_counts().head(10)
```

```python
top_mun = df["MUNICIPIO"].value_counts().head(10)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_dep.index, y=top_dep.values, palette='viridis')
ax.bar_label(ax.containers[-1], fmt='%.0f', label_type='center', rotation=45)
plt.xticks(rotation=75)
plt.title("10 primeros departamentos con más casos de violencia intrafamiliar")
plt.xlabel("Departamento")
plt.ylabel("Casos")
plt.show()

```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_mun.index, y=top_dep.values, palette='viridis')
ax.bar_label(ax.containers[-1], fmt='%.0f', label_type='center', rotation=45)
plt.xticks(rotation=75)
plt.title("10 primeros municipios con más casos de violencia intrafamiliar")
plt.xlabel("Municipio")
plt.ylabel("Casos")
plt.show()
```

## Armas Mas Comunes

```python
top_arms = df["ARMAS MEDIOS"].value_counts()
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_arms.index, y=top_arms.values, palette="viridis")
ax.bar_label(ax.containers[-1], fmt='%.0f', label_type='center', rotation=45)
plt.xticks(rotation=75)
plt.title("Armas | Medios Mas Comunes")
plt.xlabel("Armas | Medios")
plt.ylabel("Casos")
plt.show()
```

## Historico de Casos

```python
df['AÑO'] = df['FECHA HECHO'].dt.year
```

```python
plt.figure(figsize=(10,6))
ax = sns.countplot(data = df, x='AÑO', color='orange')
ax.bar_label(ax.containers[-1], fmt='%.0f', label_type='center', rotation=45)
plt.xticks(rotation=75)
plt.title("Historico de Casos")
plt.xlabel("Casos")
plt.ylabel("Año")
plt.show()
```

## Victimas por Genero

```python
# Remplazar los datos "-" por "NO REPORTA"
df['GENERO'].replace('-', 'NO REPORTA', inplace=True)

total_genero = df["GENERO"].value_counts()
```

```python
plt.pie(total_genero.values, labels=['FEMENINO', 'MASCULINO', "NO REPORTA"], autopct='%.0f%%')
plt.title("Victimas por Genero")
plt.show()
```
